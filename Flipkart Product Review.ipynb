```python
import requests
from bs4 import BeautifulSoup as bs
```

```python
url="https://www.flipkart.com/hp-15s-ryzen-3-dual-core-3250u-4-gb-1-tb-hdd-windows-10-home-15s-gr0006au-thin-light-laptop/product-reviews/itmdbb2d234797b6?pid=COMFZHZFZFFBJBHH&lid=LSTCOMFZHZFZFFBJBHH7WNMPE&marketplace=FLIPKART"
    
```

```python
r=requests.get(url)
```

```python
htmlcont=r.content
```

```python
soup=bs(htmlcont,"html.parser")
```

```python
print(soup.prettify())
```

# Title of the product


```python
t=soup.title
print(type(t))
```

```python
print(t.string)
```

# Check the Price of the Product


```python
price=soup.find_all("div",class_="_30jeq3")
```

```python
print(price)
```

```python
product_price=[]
for i in range(len(price)):
    product_price.append(price[i].get_text())
        
    
```

```python
product_price
```

# Get all the names of customer

```python
names=soup.find_all("p",class_="_2sc7ZR _2V5EHH")
names
```

```python
cust_name=[]
for i in range(len(names)):
    cust_name.append(names[i].get_text())
```

```python
cust_name
```

# Get all the Customer Comments 

```python
comment=soup.find_all("p",class_="_2-N8zT")
```

```python
comment
```

```python
cust_comm=[]
for i in range(len(comment)):
    cust_comm.append(comment[i].get_text())
```

```python
cust_comm
```

# Get Ratings of Customers

```python
rating=soup.find_all("div",class_="_3LWZlK _1BLPMq")
```

```python
rating
```

```python
cust_rating=[]
for i in range(len(rating)):
    cust_rating.append(rating[i].get_text())
```

# Get all Customer Reviews

```python
reviews=soup.find_all("div",class_="t-ZTKy")
reviews
```

```python
cust_reviews=[]
for i in range(len(reviews)):
    cust_reviews.append(reviews[i].get_text())
```

```python
cust_reviews
```

# Converting all the Data into DataFrame

```python
import pandas as pd
df=pd.DataFrame()
```

```python
df["Customer Names"]=cust_name
df["Comments"]=cust_comm
df["Stars"]=cust_rating
df["Customer Reviews"]=cust_reviews
```
